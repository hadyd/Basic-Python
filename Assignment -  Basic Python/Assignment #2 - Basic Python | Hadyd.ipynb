## Jawaban Soal 1

```python
print ("Selamat datang!")

contact_name = []
contact_number = []

def show_list_menu():
    print ("\n-- Menu --")
    print ("1. Daftar Kontak")
    print ("2. Tambah Kontak")
    print ("3. Keluar\n")
    
def add_contact():
    name = input("Nama : ")
    number = str(input("Nomer Telepon : "))
    contact_name.append(name)
    contact_number.append(number)
    print("Kontak berhasil ditambahkan")

def show_contact_list():
    print("Daftar Kontak:")
    for i in range(len(contact_name)):
        print("Nama: {}".format(contact_number[i]))
        print("Nomer Telepon: {}".format(contact_number[i]))
        
while True:
    show_list_menu()
    
    menu = int(input("Pilih menu:"))
    
    if menu == 1: 
        show_contact_list()
    elif menu == 2: 
        add_contact()
    elif menu == 3: 
        print("Program selesai, sampai jumpa!")
        break
    else: 
        print ("Menu tidak tersedia")
```

## Jawaban Soal 2

```python
def show_list_main_menu():
    print ("1. Menghitung dengan kalkulator")
    print ("2. Menemani curhatmu")
    print ("3. Keluar")

def show_menu_calculator():
    print("Menu Kalkulator")
    print("1. Penjumlahan")
    print("2. Pengurangan")
    print("3. Perkalian")
    print("4. Pembagian")
    print("5. Keluar")
    
def penjumlahan(x, y):
    return x + y

def pengurangan(x, y):
    return x - y
 
def perkalian(x, y):
    return x * y
 
def pembagian(x, y):
    return x / y  
      
print("Hai aku AIs, asisten virtual pertamamu")
print("Siapa namamu?")

nama  = str(input("Masukkan nama: " ))

default = "Maaf, aku belum paham yang kamu tanyakan"

print("\nSalam kenal {}".format(nama))
print("Apakah ada yang bisa aku bantu?")
print ("Silahkan pilih bantuan yang bisa aku berikan:")

while True:
    print(" ")
    show_list_main_menu()
    menu = int(input("Pilih menu:"))
    print(" ")

    if menu == 1:
        while True:
            print(" ")
            show_menu_calculator()
            operasi = int(input("Pilih Operasi:"))
            print(" ")
               
            x = int(input("Masukan angka pertama: "))
            y = int(input("Masukan angka ke dua: "))
            
            if operasi == 1:
                print(x , "+" , y , "=" , penjumlahan(x, y))
            elif operasi == 2:
                print(x , "-", y, "=", pengurangan(x, y))
            elif operasi == 3:
                print(x, "x", y, "=", perkalian(x, y))
            elif operasi == 4:
                print(x, ":", y, "=", pembagian(x, y))
            elif operasi == 5:
                print("Oke, terima kasih sudah menggunakan kalkulator ini")
                break
            else:
                print(default)
            break  
                
    elif menu == 2:
        input('Coba ceritakan curhatmu:')
        while True:
            Cerita = input('Coba ceritakan lebih banyak lagi\n')

            if 'bodoh' in Cerita or 'gagal' in Cerita or 'kalah' in Cerita or 'jelek' in Cerita:
                print('Kenapa kamu berpikir seperti itu?')
            elif 'kecewa' in Cerita or 'sedih' in Cerita or 'putus asa' in Cerita:
                print('Sudah berapa lama kamu merasa seperti itu?')
            elif 'senang' in Cerita or 'berhasil' in Cerita or 'menang' in Cerita or 'hadiah' in Cerita:
                input ('Itu sangat menarik!: ')
            else:
                print('\nOke, tetap semangat ya!', ':)')
                print("=====================")
                break
                
    elif menu == 3:  
        print("Terimakasih, sampai jumpa!")
        break
    else:
        print(default)
```
