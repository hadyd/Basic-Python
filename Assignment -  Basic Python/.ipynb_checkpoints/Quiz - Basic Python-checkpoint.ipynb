```python
import datetime

def show_list_menu():
    print ("1. Tentang hari ini")
    print ("2. Mencatat jadwal agenda harianmu")
    print ("3. Keluar")

def show_menu_agenda():
    print("Apa agendamu?")
    print("1. Daftar agenda")
    print("2. Menambah agenda")
    print("3. Keluar")

def show_agenda():
    print("Daftar Agenda:")
    print(" ")
    for agenda in list_agenda:
        print("Rencana kegiatanmu: " + agenda["Kegiatan"])
        print("Dilakukan tanggal: " + agenda["Tanggal"])
        print("Dilakukan jam: " + agenda["Jam"])
        print("--------------------------------------")
        print(" ")

def agenda_kosong():
    if not list_agenda:
        print("=================================")
        print ("Daftar agendamu kosong _________!")
        print ("Coba kamu tambahin agendamu dulu!")
        print("=================================")
    
list_agenda = []  

def add_agenda():
    Kegiatan = input("Apa rencana kegiatanmu: ")
    Tanggal = input("Tanggal berapa itu dilakukan (format DD-MM-YYYY): ")
    Jam = input("Dilakukan jam berapa (format HH:MM): ")
    agenda = {
        "Kegiatan": Kegiatan,
        "Tanggal": Tanggal,
        "Jam": Jam,
    }
    list_agenda.append(agenda)
    print(" ")
    print("Agendamu berhasil ditambahkan")

def namafile():
    f=open("Daftar agenda.txt", "w")
    f.write(str(list_agenda))
    f.close()
 

def show_menu_tentang_hari_ini():
    print ("Apa yang kamu ingin tahu tentang hari ini?")
    print ("1. Hari ini hari apa?")
    print ("2. Tanggal berapa hari ini?")
    print ("3. Keluar")

x = datetime.datetime.now()
tanggal = x.strftime("%d-%m-%Y")
hari = x.strftime("%A")
hari_id = {"Sunday": "Minggu","Monday": "Senin","Tuesday": "Selasa","Wednesday": "Rabu","Thursday": "Kamis",
            "Friday": "Jumat","Saturday": "Sabtu"
}
      
print("Hai aku AIs, asisten virtual pertamamu")
print("Siapa namamu?")

nama  = str(input("Masukkan nama: " ))

default = "Menu tidak tersedia"

print(" ")

print("Salam kenal {}".format(nama))
print("Apakah ada yang bisa aku bantu?")

tanggal = [
      "Ya ampun masa tidak tahu, hari ini kan hari {}".format(hari_id[hari]),
      "Hari ini tanggal {}".format(tanggal)
    ]


while True:
    print(" ")
    show_list_menu()
    menu = input("Pilih hanya angka aja ya:")
    print(" ")
            
    if menu == "2":
        while True:
            print(" ")
            show_menu_agenda()
            menu_agenda = input("Pilih hanya angka aja ya:")
            print(" ")

            if menu_agenda == "1":
                show_agenda()
                agenda_kosong()
            elif menu_agenda == "2":
                add_agenda()
            elif menu_agenda == "3":
                print("Yasuda kalo gitu")
                print("================")
                break
            else:
                print(default)

        namafile()
        
    elif menu == "1":
        #Tentang kamu dan hari ini
        while True:
            print(" ")
            show_menu_tentang_hari_ini()
            menu = input("Pilih hanya angka aja ya:")
            print(" ")

            if menu == "1": 
                print(tanggal[0])
            elif menu == "2": 
                print(tanggal[1])
            elif menu == "3": 
                print("Yasuda kalo gitu")
                break
            else: 
                print (default)  
                
    elif menu == "3":  
        print("Oke, sampai jumpa")
        break
    else:
        print(default)
```
